```python
import pandas as pd
import csv
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./loan_default_data_formatted.csv', on_bad_lines='warn')
df
```

```python
# drop unnamed column
df = df.drop(df.columns[0], axis=1)
df
```

```python
# copy to new df
loan_df = df.copy()
loan_df
```

```python
loan_df.info()
```

```python
# list out the quantity of null values in each column
for col in loan_df.columns:
    print(f"Column '{col}': {loan_df[col].isnull().sum()}")
```

```python
# display and drop rows with no values
# display and drop rows with values less than 80% 

total_row = len(loan_df)
more_than_80_percent = (total_row * 80)/100

for col in loan_df.columns:
    col_value = loan_df[col].isnull().sum()
    if col_value == total_row:
        loan_df= loan_df.drop([col], axis=1)
        print("Column with no values: ", col)
    elif col_value > more_than_80_percent:
        print("Column with values less than half: ", col)
        loan_df= loan_df.drop([col], axis=1)
```

```python
# display rows with only missing values
for col in loan_df.columns:
    sum_null = loan_df[col].isnull().sum()
    # print(sum_null)
    if sum_null > 0:
        print(f"Column '{col}': {loan_df[col].isnull().sum()}")
```

```python
# drop rows in columns that has less than 10% missing values
total_row = len(loan_df)
threshold = (total_row * 10)/100

print(threshold)
for col in loan_df.columns:
    sum_null = loan_df[col].isnull().sum()
    # print(sum_null)
    if sum_null <= threshold:
        print(f"Column '{col}': {loan_df[col].isnull().sum()}")
        loan_df = loan_df.dropna(subset=[col])
```

Handle missing values

```python
for col in loan_df.columns:
    sum_null = loan_df[col].isnull().sum()
    # print(sum_null)
    if sum_null > 0:
        print(f"Column '{col}': {loan_df[col].isnull().sum()}")
```

```python
loan_df['months_since_last_delinquency'].value_counts()
```

```python
# months since last deliquency - a situation where a borrower is late or overdue on a payment
# assume that the null values means that the borrower is never late
# replace missing value with category 'not available'

values = 'NOT AVAILABLE'
loan_df['months_since_last_delinquency'] = loan_df['months_since_last_delinquency'].fillna(value = values)
print(loan_df['months_since_last_delinquency'].value_counts())
```
